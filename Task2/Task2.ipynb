## STEP 1 : DATA CLEANING

**Step 1.1 : Remove/handle missing values (TotalCharges)**

```python
telco_customer_data['TotalCharges'].isnull().sum()
```

**Step 1.2 : Convert data types (tenure → int, charges → float)**

```python
telco_customer_data.dtypes
```

```python
telco_customer_data['TotalCharges'] = pd.to_numeric(
    telco_customer_data['TotalCharges'], 
    errors='coerce'
)
```

```python
telco_customer_data.dtypes
```

**Step : 1.3 From the feature TotalChages Drop NaN**

```python
telco_customer_data[telco_customer_data['TotalCharges'].isna()]
```

**11 rows are NaN in the feature TotalCharges**

```python
# Drop the NaN from TotalCharges
telco_customer_data.dropna(subset=['TotalCharges'], inplace=True)
```

```python
telco_customer_data
```

**Step 1.4 : Convert Churn to binary (1/0)**

```python
telco_customer_data['Churn'] = (
    telco_customer_data['Churn']
    .replace({'Yes': 1, 'No': 0})
    .infer_objects(copy=False)
)
```

```python
telco_customer_data['Churn'].head()
```

## STEP 2 : BUSINESS METRICS

**Step 2.1 : Calculating Overall Churn Rate**

```python
# Churn Rate = Customers who left / Total customers.
churn_rate = sum(telco_customer_data['Churn']) / len(telco_customer_data['Churn'])
(churn_rate * 100)
```

**Identified an overall churn rate of 26.5%, indicating approximately 1 in 4 customers discontinued service.**

**Step 2.2 : Calculating Retention Rate**

```python
# Retention Rate = 1 − Churn Rate
retention_rate = 1 - churn_rate
(retention_rate * 100)
```

**Identified an overall retention rate of 73.5%, indicating that nearly 3 out of 4 customers continued their subscription with the company.**

**Step 2.3 : Calculating Avg Tenure**

```python
avg_tenure = telco_customer_data['tenure'].mean()
avg_tenure
```

**The typical customer remains subscribed for roughly 2 years and 8 months.**

**Step 2.4 : Caculating Avg Monthly Charges**

```python
avg_monthly_charges = telco_customer_data['MonthlyCharges'].mean()
avg_monthly_charges
```

**On average, customers pay approximately $64.79 per month.**

**Step 2.5 : Calculating Avg Lifetime Value (MonthlyCharges × tenure)**

```python
avg_lifetime_value = telco_customer_data['MonthlyCharges'] * telco_customer_data['tenure']
avg_lifetime_value
```

```python
avg_lifetime_value.mean()
```

**This Means :**

**On average:**

**A customer stays ≈ 32 months**

**Pays ≈ $64.76 per month**

**Generates ≈ $2,279 in total revenue during their lifetime**

**Step 2.6 : Calculating Customer Count by Contract**

```python
telco_customer_data['Contract'].unique()
```

```python
telco_customer_data['Contract'].value_counts()
```

**55% of customers are on month-to-month contracts, showing a strong preference for flexibility and a potentially higher churn risk compared to long-term contracts.**

**Step 2.7 : Calculating Revenue by Churn vs Non-Churn**

```python
revenue_by_churn = telco_customer_data.groupby('Churn')['TotalCharges'].sum()
revenue_by_churn
```

**The majority of revenue comes from retained customers, while churned customers contributed significantly less. Reducing churn could help protect and expand long-term revenue streams.**

## STEP 3 : DEEP CHURN ANALYSIS

**Step 3.1 : Churn by Contract Type**

```python
churn_by_contract = pd.crosstab(
    telco_customer_data['Contract'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100

churn_by_contract
```

**Month-to-month customers have extremely high churn**

**Two-year contracts almost never churn**

**STRONG BUSINESS INSIGHT:**

**Contract length is the strongest predictor of churn. Encouraging long-term contracts could significantly reduce churn rate.**

**Step 3.2 : Churn by Internet Service**

```python
churn_by_internet_service = pd.crosstab(
    telco_customer_data['InternetService'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
churn_by_internet_service
```

**Customers subscribed to fiber optic services exhibit the highest churn rate (~42%), nearly double that of DSL users and significantly higher than customers without internet service. This suggests pricing sensitivity or service dissatisfaction within the fiber segment, making it a critical focus area for churn reduction strategies.**

**Step 3.3 Churn by Payment Method**

```python
churn_by_payment_method = pd.crosstab(
    telco_customer_data['PaymentMethod'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
churn_by_payment_method
```

**Customers using electronic check show higher churn rates, possibly indicating lower engagement or automatic billing adoption issues.**

**Step 3.4 Churn by Senior Citizen**

```python
churn_by_senior_citizen = pd.crosstab(
    telco_customer_data['SeniorCitizen'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
churn_by_senior_citizen
```

**Senior customers may require better service support or tailored plans.**

**Step 3.5 Churn by Tenure Buckets**

```python
telco_customer_data['TenureGroup'] = pd.cut(
    telco_customer_data['tenure'],
    bins=[0,12,24,48,60,72],
    labels=['0-12','12-24','24-48','48-60','60-72']
)
```

```python
telco_customer_data.head()
```

```python
churn_by_tenure_group = pd.crosstab(
    telco_customer_data['TenureGroup'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
churn_by_tenure_group
```

**Most churn happens in the first year of subscription. Early-stage retention strategies are critical.**

**Step 3.6 Churn by Monthly Charges Buckets**

```python
telco_customer_data['ChargesGroup'] = pd.cut(
    telco_customer_data['MonthlyCharges'],
    bins=[0,35,70,100,120],
    labels=['Low','Medium','High','Very High']
)

churn_by_monthly_charges_bucket = pd.crosstab(
    telco_customer_data['ChargesGroup'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
churn_by_monthly_charges_bucket
```

```python
telco_customer_data.head()
```

**Higher charges → higher churn**

**Customers paying premium prices may be more price-sensitive or expect higher service quality.**

## STEP 4: RETENTION & RISK INSIGHTS

**Step 4.1 : Churn by Tenure (Early Churn Detection)**

```python
# Create Tenure Buckets
telco_customer_data['TenureGroup'] = pd.cut(
    telco_customer_data['tenure'],
    bins=[0,12,24,36,48,60,72],
    labels=['0-12','12-24','24-36','36-48','48-60','60-72']
)
# Calculate Churn % by Tenure Group
churn_by_tenure = pd.crosstab(
    telco_customer_data['TenureGroup'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100

churn_by_tenure
```

**Nearly half of customers churn within the first year, while long-term customers show very low churn, making early retention efforts critical.**

**Step 4.2 : High-Value Customer Churn**

```python
# Create CLV (Customer Lifetime Value)
telco_customer_data['CLV'] = (
    telco_customer_data['MonthlyCharges'] *
    telco_customer_data['tenure']
)
# Compare Average CLV by Churn
telco_customer_data.groupby('Churn')['CLV'].mean()
# Revenue at Risk (Total Revenue from Churned Customers)
telco_customer_data.loc[
    telco_customer_data['Churn'] == 1,
    'TotalCharges'
].sum()
```

**Churned customers account for $2.86M in lost revenue, highlighting significant financial impact and the importance of reducing churn.**

**Step 4.3 : Long-Term vs Short-Term Customer Stability**


```python
# Create Stability Category
telco_customer_data['CustomerType'] = telco_customer_data['tenure'].apply(
    lambda x: 'Short-Term' if x <= 24 else 'Long-Term'
)
# Churn % by Customer Type
pd.crosstab(
    telco_customer_data['CustomerType'],
    telco_customer_data['Churn'],
    normalize='index'
) * 100
```

**Short-term customers are nearly 3x more likely to churn than long-term customers, making early retention a critical business priority.**
